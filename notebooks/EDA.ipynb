## Columns Info
- ID: A unique identifier for each entry in the dataset.
- Delivery_person_ID: A unique identifier for each delivery person.
- Delivery_person_Age: Age of the delivery person.
- Delivery_person_Ratings: Ratings or performance score of the delivery person.
- Restaurant_latitude: Latitude coordinates of the restaurant's location.
- Restaurant_longitude: Longitude coordinates of the restaurant's location.
- Delivery_location_latitude: Latitude coordinates of the delivery location.
- Delivery_location_longitude: Longitude coordinates of the delivery location.
- Order_Date: Date when the order was placed.
- Time_Ordered: Time when the order was placed.
- Time_Order_picked: Time when the order was picked up by the delivery person.
- Weather_conditions: Weather conditions during the delivery (e.g., sunny, rainy).
- Road_traffic_density: Traffic density on the roads during the delivery (e.g., low, medium, high, jam).
- Vehicle_condition: Condition of the delivery vehicle.
- Type_of_order: Type of order (e.g., Drink ,snack).
- Type_of_vehicle: Type of delivery vehicle used (e.g., bike, car, scooter).
- Multiple_deliveries: Whether the delivery person handled multiple deliveries during the trip.
- Festival: Whether the delivery was made during a festival or special event.
- City: City where the delivery was made.
- Time_taken (min): Time taken for the delivery in minutes (target variable).


```python
## So from columns we can see that the 'ID' and 'Delivery_person_ID' will not be much of use so dropping them
## also creating a copy of original dataset before making any changes

df_copy = df.copy()
df_copy.drop(labels=['ID', 'Delivery_person_ID'] ,axis=1 ,inplace=True)
```

```python
# checking the information about a DataFrame 
df_copy.info()


## the non-null value is varying so there might be null values in multiple columns
## 'Order_Date', 'Time_Orderd' and'Time_Order_picked' are all cat columns
## As the Time columns are categorical then it has ':' symbol 
```

```python
## Check for null values
df_copy.isna().sum()

## Multiple columns has null value
```

```python
## checking for duplicated
df_copy.duplicated().sum()

## no duplicate found
```

```python
## Checking the value counts in all the categorical columns
#  excluding date and time columns 
exclude_columns = ['Order_Date', 'Time_Orderd', 'Time_Order_picked']


columns = df_copy.select_dtypes(include='object').columns.to_list()
columns = [col for col in columns if col not in exclude_columns]


for i in columns:
    print(df_copy[i].value_counts())
    print("*" * 39)
```

## From the value counts observations:

- The most common weather condition is "Fog," and the least common is "Sunny."
- The majority of road traffic densities are  "Low". " "High" road traffic density is the least common.
- "Motorcycle" is the most frequently used type of vehicle. "Bicycle" is the least used type of vehicle.
- Most of the deliveries were not during a festival ("No") .
- The majority of deliveries were made in "Metropolitan" cities.
- Also `Type_of_order` would not be useful feature so droping it

```python
df_copy.drop(labels=['Type_of_order'] ,axis=1 ,inplace=True)
```

```python
## Lets deal with the null values and then proceed 
# Fill missing values in numerical columns with median
numerical_columns = df_copy.select_dtypes(include=np.number).columns
for col in numerical_columns:
    median_value = df_copy[col].median()
    df_copy[col].fillna(median_value, inplace=True)

# Fill missing values in categorical columns with most frequent value
categorical_columns = df_copy.select_dtypes(include='object').columns
for col in categorical_columns:
    most_frequent_value = df_copy[col].mode().iloc[0]
    df_copy[col].fillna(most_frequent_value, inplace=True)
```

```python
# Grouping by 'Delivery_person_Ratings' and calculating avg 'Time_taken (min)'
ratings_vs_time = df_copy.groupby('Delivery_person_Ratings')['Time_taken (min)'].mean()
print("Ratings vs. Time Taken (min):\n", ratings_vs_time)

# Ratings vs. Time Taken (min) - Scatter Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Delivery_person_Ratings', y='Time_taken (min)', data=df_copy)
plt.title('Ratings vs. Time Taken (min)')
plt.xlabel('Delivery Person Ratings')
plt.ylabel('Time Taken (min)')
plt.show()


#delivery person who have higher rating above 4.5 made delivery within 25mins on an average
```

```python
# Grouping by 'City' and calculating avg  'Time_taken (min)' 
city_vs_time = df_copy.groupby('City')['Time_taken (min)'].mean()
print("\nCity vs. Time Taken (min):\n", city_vs_time)

# City vs. Time Taken (min) - Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Time_taken (min)', data=df_copy)
plt.title('City vs. Time Taken (min)')
plt.xlabel('City')
plt.ylabel('Time Taken (min)')
plt.show()

##Urban areas  have the lowest delivery times, while Semi-Urban areas have the highest delivery times.
##Urban areas and Metropolitian  areas are generally more developed and have better road infrastructure, which allows delivery persons to navigate more efficiently and deliver orders quickly. On the other hand, Semi-Urban areas might have less developed infrastructure and more challenging traffic conditions, resulting in longer delivery times.
##City value count
# Metropolitian    34087
# Urban            10133
# Semi-Urban         164
```

```python
# Grouping by 'City' and 'Road_traffic_density', and getting most frequent rating for 'Delivery_person_Ratings'
city_traffic_vs_ratings = df_copy.groupby(['City', 'Road_traffic_density'])['Delivery_person_Ratings'].apply(lambda x: x.mode().iloc[0])
print("\nCity & Traffic Density vs. Ratings:\n", city_traffic_vs_ratings)

# City & Traffic Density vs. Ratings - Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Delivery_person_Ratings', hue='Road_traffic_density', data=df_copy)
plt.title('City & Traffic Density vs. Ratings')
plt.xlabel('City')
plt.ylabel('Delivery Person Ratings')
plt.legend(title='Traffic Density')
plt.show()

## The rating is mostly consistent so cant say much
```

```python
# Grouping by 'City' and 'Road_traffic_density', and calculating avg 'Time_taken (min)'
city_traffic_vs_time = df_copy.groupby(['City', 'Road_traffic_density'])['Time_taken (min)'].mean()
print("\nCity & Traffic Density vs. Time_taken:\n", city_traffic_vs_time)

# City & Traffic Density vs. Time_taken - Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Time_taken (min)', hue='Road_traffic_density', data=df_copy)
plt.title('City & Traffic Density vs. Time Taken')
plt.xlabel('City')
plt.ylabel('Time Taken (min)')
plt.legend(title='Traffic Density')
plt.show()

# The time taken for deliveries is influenced by both city and traffic density. In Metropolitan areas, high traffic density leads to longer delivery times, while low traffic density results in shorter delivery times. In Semi-Urban areas, traffic density does not seem to have a significant impact on delivery times.In urban area the delivery time is pretty less compared to other cities
```

```python
# Grouping by 'Delivery_person_Age' and calculating mean of 'Delivery_person_Ratings'
age_vs_ratings = df_copy.groupby('Delivery_person_Age')['Delivery_person_Ratings'].mean()
print("\nAge vs. Ratings:\n", age_vs_ratings)

# Age vs. Ratings - Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Delivery_person_Age', y='Delivery_person_Ratings', data=df_copy)
plt.title('Age vs. Ratings')
plt.xlabel('Delivery Person Age')
plt.ylabel('Delivery Person Ratings')
plt.show()

##Older delivery persons have slightly higher ratings, while younger ones have slightly lower ratings.These could be because older delivery persons may have more experience in the job, which could lead to better skills in handling deliveries and navigating through different traffic conditions
```

```python
# Grouping by 'Delivery_person_Age' and 'City', and calculating mean of 'Time_taken (min)'
age_city_vs_time = df_copy.groupby(['Delivery_person_Age', 'City'])['Time_taken (min)'].mean()
print("\nAge & City vs. Time Taken (min):\n", age_city_vs_time)

# Age & City vs. Time Taken (min) - Bar Plot
plt.figure(figsize=(15, 10))
sns.barplot(x='Delivery_person_Age', y='Time_taken (min)', hue='City', data=df_copy)
plt.title('Age & City vs. Time Taken (min)')
plt.xlabel('Delivery Person Age')
plt.ylabel('Time Taken (min)')
plt.legend(title='City')
plt.show()

# The age of the delivery person appears to have some impact on delivery times in different cities. In Metropolitan areas, younger delivery persons have longer delivery times, while in Urban areas, younger delivery persons have shorter delivery times. In Semi-Urban areas, delivery times seem to be consistent across different age groups.

##These could be because they might still be learning the most efficient routes and dealing with complex traffic conditions as big city tends to have various routes for each location. In Urban areas, younger delivery persons might have shorter delivery times due to better familiarity with the locality and traffic patterns.
```

```python
# Grouping by 'Traffic_density'  and calculating mean of 'Time_taken (min)'
traffic_vs_time = df_copy.groupby(['Road_traffic_density'])['Time_taken (min)'].mean()
print("\nTraffic  vs. Time Taken (min):\n", traffic_vs_time)

# Traffic vs. Time Taken (min) - Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Road_traffic_density', y='Time_taken (min)', data=df_copy)
plt.title('Traffic vs. Time Taken (min)')
plt.xlabel('Traffic Density')
plt.ylabel('Time Taken (min)')
plt.show()
# Traffic density is directly related to the time taken for deliveries. High traffic density is associated with longer delivery times, while low traffic density is linked to shorter delivery times.
```

```python
# Grouping Weather against Time Taken
weather_vs_time = df_copy.groupby('Weather_conditions')['Time_taken (min)'].mean()
print("Weather vs. Time Taken (min):\n", weather_vs_time)

# Weather vs. Time Taken - Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Weather_conditions', y='Time_taken (min)', data=df_copy)
plt.title('Weather vs. Time Taken (min)')
plt.xlabel('Weather Conditions')
plt.ylabel('Time Taken (min)')
plt.show()

#The time taken for deliveries is influenced by the weather conditions. Deliveries tend to take the longest during foggy and cloudy weather conditions, while they are the shortest during sunny weather.
# most suitable condition would be sunny 
```

```python
#  Grouping Type of Vehicle against Time Taken
vehicle_vs_time = df_copy.groupby('Type_of_vehicle')['Time_taken (min)'].mean()
print("\nType of Vehicle vs. Time Taken (min):\n", vehicle_vs_time)

# Type of Vehicle vs. Time Taken - Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Type_of_vehicle', y='Time_taken (min)', data=df_copy)
plt.title('Type of Vehicle vs. Time Taken (min)')
plt.xlabel('Type of Vehicle')
plt.ylabel('Time Taken (min)')
plt.show()

# Electric scooters and scooters tend to have the shortest delivery times, while bicycles have slightly longer delivery times. Motorcycles have the longest delivery times among the different vehicle types.
```

## Summary of Exploratory Data Analysis (EDA):

1. Data Overview:
   - We observed that there are some missing values in the dataset, which were handled by filling them with appropriate values.
   - The 'ID' and 'Delivery_person_ID' columns were dropped as they were unique identifiers and didn't provide any meaningful information for analysis.
   - The time columns have fraction values which we need  to take care.

2. Delivery Time vs. Delivery Person Ratings:
   - Delivery persons with higher ratings (above 4.5) tend to complete deliveries within 25 minutes on average.


3. Delivery Time in Different Cities:
   - Deliveries in Urban areas have the shortest delivery times, while Semi-Urban areas have the longest delivery times.
   - Metropolitan cities have the highest number of deliveries, followed by Urban and Semi-Urban areas.

4. Delivery Time vs. Traffic Density:
   - Traffic density is directly related to delivery time. Higher traffic density leads to longer delivery times, while lower traffic density results in shorter delivery times.

5. Delivery Time vs. Weather Conditions:
   - Deliveries take the longest during foggy and cloudy weather conditions.
   - Sunny weather conditions have the shortest delivery times.

6. Delivery Time vs. Type of Vehicle:
   - Electric scooters and scooters have the shortest delivery times, while motorcycles have the longest delivery times.

7. Delivery Time vs. Multiple Deliveries:
   - As the number of multiple deliveries increases, the delivery time also increases significantly. More stops and additional drop-offs require more time for each delivery.



## Conclusion:
   - Delivery person ratings, traffic density, weather conditions, multiple deliveries and type of vehicle are significant factors that influence delivery times.
   - Delivery times tend to be shorter in Urban and Metropolitan areas with lower traffic density and sunny weather conditions.
   - Festivals have a significant impact on delivery times, leading to longer delivery durations.
   - During cloudy and foggy weather conditions, delivery times are generally longer, especially in areas with high traffic density. 
   - Older delivery persons, possibly with more experience in the job, tend to have slightly higher ratings. This may imply that experienced delivery persons have better skills in handling deliveries and navigating through different traffic conditions.
